```python
import easyocr
import cv2
import matplotlib.pyplot as plot
import numpy as np
```

```python
IMG_P= '/Users/mac/Desktop/ejemplo_4_ocr.jpg'
reader = easyocr.Reader(['en'])
RST = reader.readtext(IMG_P)
RST
```

```python
T_LEFT = tuple(RST[0][0][0])
B_RIGHT = tuple(RST[0][0][2])
text = RST[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
IMG = cv2.imread('/Users/mac/Desktop/ejemplo_1_ocr.png')
IMG = cv2.rectangle(IMG,T_LEFT,B_RIGHT,(0,255,0),3)
IMG = cv2.putText(IMG,text,T_LEFT, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plot.imshow(IMG)
plot.show()
```

```python
IMG = cv2.imread(IMG_P)
spacer = 100
for detection in RST:
    T_LEFT = tuple(detection[0][0])
    B_RIGHT = tuple(detection[0][2])
    TEXT = detection[1]
    IMG = cv2.rectangle(IMG,T_LEFT,B_RIGHT,(0,255,0),3)
    IMG = cv2.putText(IMG,TEXT,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
```

```python
plot.imshow(IMG)
plot.show()
```
